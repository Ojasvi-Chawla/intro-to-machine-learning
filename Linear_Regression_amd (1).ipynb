# Detecting Outliers

```python
columns = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

all_outlier_indices = set()  # To store unique outlier row indices

# IQR outlier check per column
for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # Detect outliers
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    outlier_indices = outliers.index.tolist()
    all_outlier_indices.update(outlier_indices)
    
    total = len(df[col])
    outlier_count = len(outliers)
    outlier_percent = (outlier_count / total) * 100
    
    print(f"\nChecking column: {col}")
    if outlier_count > 0:
        print("Outliers detected:")
        print(outliers)
        print(f"Number of outliers: {outlier_count}")
        print(f"Percentage of outliers: {outlier_percent:.2f}%")
    else:
        print("No outlier detected.")

# Calculate total outlier rows (unique)
total_rows = len(df)
total_outliers = len(all_outlier_indices)
total_outlier_percent = (total_outliers / total_rows) * 100

print(f"\n Total number of unique outlier rows in the dataset: {total_outliers}")
print(f" Overall percentage of containing outlier: {total_outlier_percent:.2f}%")
```
